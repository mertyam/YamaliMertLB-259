## Vorhersageziel
Ich mÃ¶chte das Close-Feld vorhersagen, da es den letzten gehandelten Preis darstellt. Es ist ein zentraler Referenzwert fÃ¼r HÃ¤ndler und wird in der technischen Analyse oft verwendet.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Daten laden
df = pd.read_csv("C:/Users/Mert/OneDrive - BBBaden/BBB/259/XRP_Data.csv")
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Spalten, die skaliert werden sollen (Eingabedaten)
features_to_scale = ['Open', 'High', 'Low', 'Volume']
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df[features_to_scale])

# Skalierten DataFrame erstellen mit neuen Spaltennamen
scaled_df = pd.DataFrame(scaled_features, columns=[f + '_scaled' for f in features_to_scale])

# Skalierten DataFrame an Original anhÃ¤ngen
df = pd.concat([df, scaled_df], axis=1)

# Ausgabe skalierten Werte zur Kontrolle
print("\nðŸ”§ Skalierten Werte (erste 5 Zeilen):")
print(df[[f + '_scaled' for f in features_to_scale]].head())

# Erweiterte Kennzahlen fÃ¼r jedes Original-Feld (nicht skaliert)
print("\nðŸ“Š Statistische Kennzahlen (Originalwerte):")
for col in ['Open', 'Close', 'High', 'Low', 'Volume']:
    print(f"\n--- {col} ---")
    print(f"Mittelwert:            {df[col].mean():.6f}")
    print(f"Median:                {df[col].median():.6f}")
    print(f"Modus:                 {df[col].mode().iloc[0]:.6f}")
    print(f"Standardabweichung:    {df[col].std():.6f}")
    print(f"Varianz:               {df[col].var():.6f}")
    print(f"25%-Perzentil:         {df[col].quantile(0.25):.6f}")
    print(f"50%-Perzentil (Median):{df[col].quantile(0.50):.6f}")
    print(f"75%-Perzentil:         {df[col].quantile(0.75):.6f}")
```

## Visualisierungen
Histogramm: Verteilung des Close-Preises

Zeitreihe: Verlauf des Close-Preises Ã¼ber den Zeitstempel

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten laden
df = pd.read_csv("C:/Users/Mert/OneDrive - BBBaden/BBB/259/XRP_Data.csv")
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Ziel-Feld: Close
ziel_feld = 'Close'

# Histogramm fÃ¼r Close
plt.figure(figsize=(8, 4))
plt.hist(df[ziel_feld], bins=30, edgecolor='black')
plt.title('Histogramm des Close-Preises')
plt.xlabel('Close-Preis')
plt.ylabel('HÃ¤ufigkeit')
plt.grid(True)
plt.show()

# Streudiagramm: Timestamp vs. Close
plt.figure(figsize=(10, 4))
plt.plot(df['Timestamp'], df['Close'], linewidth=1)
plt.title('Close-Preis Ã¼ber die Zeit')
plt.xlabel('Zeitpunkt')
plt.ylabel('Close-Preis')
plt.grid(True)
plt.tight_layout()
plt.show()
```
